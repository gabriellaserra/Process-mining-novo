```python
dfBaseNT2023 = pandas.read_excel("dudaArqs\\Relatorio_NT_2022.xlsx")
dfBaseNT2022 = pandas.read_excel("dudaArqs\\Relatorio_NT_2023.xlsx")

# Concatenar as duas bases
dfConcatenado = pandas.concat([dfBaseNT2022, dfBaseNT2023])

# Tirar duplicatas
dfConcatenado = dfConcatenado.drop_duplicates()   ##Está tirando todas as duplicatas, queremos manter uma de cada caso(FEITO)


# Tirar colunas desnecessárias
lColunasDel = ['Classe do Navio', 'Rebocador', 'Hora Top', 'Alívio Crítico', 'Apuração Alívio Crítico', 'Volume Acumulado Perdas', 'Lifter']
dfConcatenado = dfConcatenado.drop(columns=lColunasDel)

# Padronizar o formato de data e hora
lFormatDate = ['Data Prevista de Chegada', 'Data Prevista de Saída', 
'H.O.C.', 'Amarração','Início de Bombeio', 'Fim de Bombeio', 'Desamarração', 'H.O.S.', 'Início Primeira Amarração','Fim Primeira Amarração','Início Primeira Conexão','Fim Primeira Conexão',
'Início Primeiro Bombeio','Fim Último Bombeio','Início Última Desconexão','Fim Última Desconexão','Início Última Desamarração','Fim Última Desamarração']

for coluna in lFormatDate:
    dfConcatenado[coluna] = pandas.to_datetime(dfConcatenado[coluna], dayfirst=True)

# Retirando os acentos e ç's
dfConcatenado = dfConcatenado.map(lambda x: unidecode(str(x)) if isinstance(x, str) else x)
dfConcatenado.to_excel(('dudaArqs\\Relatorios_concatenados.xlsx'), index=False)
# dfConcatenado

dfConcatenadoBase = pandas.read_excel("dudaArqs\\Relatorios_concatenados.xlsx")
dfConcatenadoBase
```
